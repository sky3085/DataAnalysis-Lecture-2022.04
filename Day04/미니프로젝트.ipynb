```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
#1-1
cctv=pd.read_csv('cctv.csv', skiprows=1, encoding='EUC-KR')
cctv.head()
```

```python
#1-2
cctv.isna().sum().sum()#Nan값이 2개 있음
```

```python
cctv.fillna(0, inplace=True)
cctv.head()
```

```python
#1-3
cctv['총계']=cctv['총계'].str.replace(',','')
cctv.head()
```

```python
#1-3
cctv.loc[0]=cctv.loc[0].str.replace(',','')
cctv.head()
```

```python
#1-4
cctv['최근증가율'] = cctv.loc[:,'2018년': '2020년'].astype(float).sum(axis=1) / cctv.loc[:, '2012년 이전': '2017년'].astype(float).sum(axis=1) *100
cctv.head()
```

```python
#1-5, 1-6
cctv_new = cctv[['구분', '총계', '최근증가율']]
cctv_new.rename(columns={'구분':'구별', '총계':'cctv댓수'}, inplace=True)
cctv_new.head()
```

```python
#1-7, 1-8
cctv_new['구별'] = cctv_new['구별'].str.replace(' ','')
cctv_new.drop([0], inplace=True)
cctv_new.head()
```

```python
#2-1
person = pd.read_csv('서울시인구.txt', sep='\t', skiprows=2)
person.head()
```

```python
#2-2
person.isna().sum().sum()
```

```python
#2-3
person_new = person[['자치구', '계', '계.1','계.2','65세이상고령자']]
person_new.head()
```

```python
#2-5
person_new['계']=person_new['계'].str.replace(',','')
person_new['계.1']=person_new['계.1'].str.replace(',','')
person_new['계.2']=person_new['계.2'].str.replace(',','')
person_new['65세이상고령자']=person_new['65세이상고령자'].str.replace(',','')
person_new.head()
```

```python
#2-4
person_new.columns=['구별','인구수','내국인','외국인','노령자']
person_new.head()
```

```python
#2-6
person_new.drop([0], inplace=True)
person_new.head()
```

```python
#2-7, 2-8
person_new['외국인비율']=person_new['외국인'].astype(float)/person_new['인구수'].astype(float)*100
person_new['노령자비율']=person_new['노령자'].astype(float)/person_new['인구수'].astype(float)*100
person_new['구별']=person_new['구별'].replace(' ','')
person_new.head()
```

```python
# 3-1,3-2
answer = pd.merge(cctv_new,person_new,how='inner')
answer.set_index('구별',inplace=True)
answer
```

```python
#3-3 #3-4
answer.to_csv('cctv_pop.csv')
an = pd.read_csv('cctv_pop.csv')
an
```
