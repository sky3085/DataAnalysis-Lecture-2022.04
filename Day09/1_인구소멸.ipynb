### 인구소멸 위기지역 - 전처리(Part1)
- 국가통계포털 : https://kosis.kr/index/index.do
- 공공데이터포털 : https://www.data.go.kr/

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/시군구_성_연령_5세_별_인구2_2020.csv', encoding='euc-kr')
print(df.shape)
df.head(260).tail(5)
```

```python
# 광역시도 시군구 20-39 65이상 소멸비율 소멸위기지역
# 시군구에 행정구(값)가 있으면 시군구 값을 행정구 값으로 변경
df.시군구[df.행정구.notnull()] = df.행정구[df.행정구.notnull()]
df.head(260).tail(5)
```

```python
# 소계 데이터 지우기 (Filtering)
df= df[df.시군구 != '소계']
print(df.shape)
df.head(260).tail(5)
```

```python
# 20-39세, 65세 이상 컬럼 만들기
df['20-39세']=df['20 - 24세']+df['35 - 39세']+df['25 - 29세']+df['30 - 34세']
df['65세이상']=df['65 - 69세']+df['70 - 74세']+df['75 - 79세']+df['80세 이상']
```

```python
df = df[['광역시도', '시군구', '구분', '인구수', '20-39세', '65세이상']]
df.head()
```

```python
# pivot talbe로
pop = pd.pivot_table(df, index=['광역시도', '시군구'], columns=['구분'],values=['인구수', '20-39세', '65세이상'])
pop.head()
```

```python
# 인구 소멸 비율 계산
pop['소멸비율'] = round(pop['20-39세', '여자']/pop['65세이상', '계'], 4)
pop.head()
```

```python
# 인구 소멸위기지역, 소멸위기 고위험지역
pop['소멸위기지역']=pop.소멸비율 < 0.5
pop['소멸위기고위험지역']=pop.소멸비율 < 0.2
pop.head()
```

```python
# 인구 소멸위기 지역
crisis_region = pop[pop.소멸위기지역].index.get_level_values(1)
print(crisis_region, len(crisis_region))
```

```python
# 인구 소멸 고위험 지역
high_crisis_region = pop[pop.소멸위기고위험지역].index.get_level_values(1)
print(high_crisis_region, len(high_crisis_region))
```

```python
# 컬럼명 변경: 이중인덱스 -> 단일 인덱스
# 65세이상 남, 여, 인구수 남, 여 제외
pop.columns.get_level_values(0)
```

```python
pop.columns.get_level_values(1)
```

```python
new_columns = []
for i in range(len(pop.columns.get_level_values(0))):
    new_columns.append(pop.columns.get_level_values(0)[i]+pop.columns.get_level_values(1)[i])
pop.columns = new_columns
pop.head()
```

```python
pop=pop.drop(columns=['65세이상남자','65세이상여자','인구수남자','인구수여자'])
pop
```

```python
#인덱스 제거
pop.reset_index(inplace=True)
pop.head(3)
```

```python
# 시군구 공백 제거
pop.시군구 = pop.시군구.str.strip()
pop.head()
```

```python
# 파일 저장
pop.to_csv('data/시군구_전처리완료.csv', index=False, encoding='euc-kr')
```
