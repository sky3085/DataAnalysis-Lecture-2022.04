```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

- Stopwords 처리

```python
# NLTK stopwords
from nltk.corpus import stopwords
sw_nltk = stopwords.words('english')
print(sw_nltk)
```

```python
# WordCloud 에서 제고하는 불용어(stopwords)
sw_wc = STOPWORDS
print(sw_wc)
```

```python
len(sw_nltk), len(sw_wc)
```

- Alice in Wonderland

```python
with open ('data/Alice.txt') as file:
    text = file.read()
```

```python
# stopwords에 'said' 단어 추가
sw_wc.add('said')
```

```python
wc = WordCloud(background_color='white', max_words=2000, stopwords=sw_wc)
wc = wc.generate(text)
```

```python
wc.words_
```

```python
count=0
for key, val in wc.words_.items():
    print(key, ":", val)
    count += 1
    if count > 10:
        break
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

- 그림 위에 표시하기

```python
mask = np.array(Image.open('data/Alice_mask.png'))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', mask=mask, max_words=1000, stopwords=sw_wc).generate(text)
```

```python
plt.figure(figsize=(10,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

- Star Wars

```python
with open('data/A_new_hope.txt') as file:
    text = file.read()
text = text.replace('HAN', 'Han').replace("LUKE'S", 'Luke')
```

```python
from wordcloud import STOPWORDS
sw = STOPWORDS
sw.add('int')
sw.add('ext')
len(sw)
```

```python
mask = np.array(Image.open('data/Stormtrooper_mask.png'))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(mask=mask, max_words=1000, stopwords=sw, margin=10, random_state=2022).generate(text)
default_colors = wc.to_array
```

```python
def grey_color(word, font_size, position, orientation, random_state, **kwargs):
    return f'hsl(0, 0%, {np.random.randint(60,101)}%)'
```

```python
plt.figure(figsize=(10,12))
plt.imshow(wc.recolor(color_func=grey_color, random_state=2022), interpolation='bilinear')
plt.axis('off')
plt.show()
```
