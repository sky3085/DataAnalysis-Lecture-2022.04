### 지니 top200 차트

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
melon_top200=[]
rank = 1
for i in range(1,5):
    url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220414&hh=15&rtm=Y&pg='
    header={'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36'}
    req = requests.get(url+f'{i}', headers=header)
    soup = BeautifulSoup(req.text, 'html.parser')

    for info in soup.select('td.info'):
        a=info.select_one('.title.ellipsis').text.strip()
        if('19금' in str(a)):
            melon_top200.append({'rank':rank, 'title':a.split(' ')[-1], 'artist':info.select_one('.artist.ellipsis').text.strip()})
            rank+=1
        else :
            melon_top200.append({'rank':rank,'title':info.select_one('.title.ellipsis').text.strip(), 'artist':info.select_one('.artist.ellipsis').text.strip()})
            rank+=1
    
    
df=pd.DataFrame(melon_top200)
df.set_index('rank', inplace=True)
df.to_csv('melon200.csv')
a = pd.read_csv('melon200.csv')
a
```
