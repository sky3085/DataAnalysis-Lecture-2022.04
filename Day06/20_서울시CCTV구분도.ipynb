### 서울시 CCTV 구별 단계구분도

```python
import json, folium
import pandas as pd
```

```python
df = pd.read_csv('cctv_pop.csv', index_col='구별')
df.head()
```

```python
file_name='json_practice.json'
geo_str = json.load(open(file_name,encoding='utf-8'))
geo_str
```

- 단계구분도

```python
map=folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_str,
    data=df.cctv댓수,
    columns=[df.index, df.cctv댓수],
    fill_color='PuRd',
    key_on='feature.id'
).add_to(map)
map
```

- 구 이름 표시하기(좌표, 텍스트)

```python
import numpy as np
def get_text_location(geo_str):
    gu_dict={}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            gu_dict[gu['id']] = [np.mean(geo[:,1]),np.mean(geo[:,0])]#(np.max(geo[:,1])+np.min(geo[:,1]))/2, (np.max(geo[:,0])+np.min(geo[:,0]))/2
    return gu_dict
```

```python
from folium.features import DivIcon
gu_dict = get_text_location(geo_str)

map=folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_str,
    data=df.cctv댓수,
    columns=[df.index, df.cctv댓수],
    fill_color='PuRd',
    key_on='feature.id'
).add_to(map)

for gu_name in df.index:
    folium.Marker(
        location=gu_dict[gu_name],
        icon=DivIcon(icon_size=(80,20), html=f'<div style="font-size:10pt; text-align:center; font-weight:bold">{gu_name}</div>')
    ).add_to(map)

map
```
