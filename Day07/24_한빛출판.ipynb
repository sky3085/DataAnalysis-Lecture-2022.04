### 한빛출판네트워크 - 새로나온 책 10 페이지

```python
from bs4 import BeautifulSoup
import pandas as pd
import re
import requests
```

```python
lines = []

for page in range(1,11):
    url = f'https://www.hanbit.co.kr/store/books/new_book_list.html?page={page}&brand=&cate1=&cate2=&searchKey=&keyWord='
    req = requests.get(url)
    soup=BeautifulSoup(req.text, 'html.parser')
    lis = soup.select('.sub_book_list_area > li')
    #print(lis)
    for li in lis:
        site='https://www.hanbit.co.kr'
        sub_href=li.select_one('.book_tit > a')['href']
        title=li.select_one('.book_tit > a').get_text()
        author = li.select_one('.book_writer').get_text()
        price=li.select_one('.price').get_text().replace(',','')
        req2=requests.get(site+sub_href)
        sub_soup=BeautifulSoup(req2.text, 'html.parser')
        sub_lis=sub_soup.select('.info_list > li')
        date='-'
        trans='-'
        for l in sub_lis:
            #출간일 쪽수 가격 역자
            if '출간 :' in l.getText().strip() :
                date=l.get_text().split(':')[1].strip()
            elif '페이지 :' in l.getText().strip():
                page=l.get_text().split(':')[-1].split("쪽")[0].strip()
            elif '번역 :' in l.getText().strip():
                trans=l.get_text().split(':')[1].strip()
        lines.append([title, author, trans, date, price, page])


df = pd.DataFrame(lines, columns=['title', 'author', 'trans','date', 'price', 'page'])
df
            
            
```
