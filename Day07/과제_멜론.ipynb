```python
from bs4 import BeautifulSoup
import pandas as pd
import re
import requests
```

```python
#1 멜론차트 Top100

#1.1 데이터 가져오기
url = 'https://www.melon.com/chart/index.htm'
header={'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36'}
req = requests.get(url, headers=header)
soup = BeautifulSoup(req.text, 'html.parser')

#1.2 원하는 데이터 찾기
lis = soup.select('tr.lst50')

lines=[]

#1.4 반복문을 사용하여 페이지에 있는 데이터 가져오기 및 오류 수정하기
for li in lis:

    # 1.3 하나를 선택해서 원하는 정보 추출하기
    title = li.select_one('.ellipsis.rank01 a').get_text()
    artist = li.select_one('.ellipsis.rank02 > a').get_text()
    rank = li.select_one('span.rank').get_text()
    album = li.select_one('.ellipsis.rank03 > a').get_text()
    lines.append([rank, title, artist, album])

#1.5 데이터프레임으로 만들고 저장하기
df = pd.DataFrame(lines, columns=['rank', 'title', 'artist', 'album'])
df.to_csv('MelonTop100.csv', index=False)
answer=pd.read_csv('MelonTop100.csv')
answer.head()
```
